```python
from imdb import IMDb
import pandas as pd
#import numpy as np
from datetime import datetime

# create an instance of the IMDb class
ia = IMDb()
```

```python
def get_serie(id_lista): #id_lista eh uma lista de ids
  colunas = {'Titulo':[],
             'Check':[],
             'IMDBid':[],
             'Nota':[],
             'Votos':[],
             'Elenco':[],
         'Gênero':[],
         'Duração':[],
         'Reviews':[],
         #'Prêmios':[],
         #'Sinopse':[],
         'Ano':[],
         'Data de estréia':[],
         #'Data dos episodios':[],
         #'Diretores':[],
         'Roteiristas':[],
         #'Produtoras':[],
         'Censura':[],
         'País':[],
         'Idiomas':[],
         'Episodios':[]}
  df = pd.DataFrame(colunas)

  for id in id_lista:
    serie = ia.get_movie(id)
    print(serie)
    titulo = serie['title']
    checagem = serie['kind']
    nota = serie['rating']
    votos = serie['votes']
    ### elenco
    elenco = serie['cast'] 
    lista_elenco = []
    for i in elenco:
      lista_elenco.append(i['name'])
    ###
    genero = serie['genres']
    try:  
      duracao = serie['runtimes'] #precisamos acessar por episodio?
    except KeyError:
      duracao = ["NA"]
    ### reviews
    ia.update(serie, 'reviews')
    reviews = len(serie['reviews'])
    ###
    #premios = serie[''] #nao ta achando
    #sinopse = serie['plot outline'] #ou só ['plot']?
    ano = serie['year']
    ##### data de estreia (primeiro episodio)
    ia.update(serie, 'episodes')
    episodes = serie.data['episodes']
    epi1 = serie['episodes'][1][1]
    #cont = 0
    #tam=10
    #while((tam<=11) and (cont<=len(serie['episodes'][1])) ):
    #  data_estreia = ia.get_movie_release_dates(epi1.movieID)['data']['raw release dates'][cont]['date']
    #  cont = cont+1
    flag = False
    for epi in ia.get_movie_release_dates(epi1.movieID)['data']['raw release dates']:
      if (epi['date'][0].isnumeric()) and (flag==False):
        data_estreia = epi['date']
        flag = True
    data_estreia = datetime.strptime(data_estreia, "%d %B %Y")
    data_estreia = data_estreia.strftime('%Y-%m-%d')
    #####
    #
    ##### data dos episodios
    ia.update(serie, 'episodes')
    episodes = serie.data['episodes']
    tamanho = len(episodes[1])
    #lista_data_epi = []
    #for i in range(1, tamanho):
    #  try:
    #    epi = episodes[1][i]
    #    data_epi = ia.get_movie_release_dates(epi.movieID)['data']['raw release dates'][0]['date']
    #  except ValueError:
    #    epi = episodes[1][1]
    #    data_epi = ia.get_movie_release_dates(epi.movieID)['data']['raw release dates'][0]['date']
    #  data_epi = datetime.strptime(data_epi, "%d %B %Y")
    #  data_epi = data_epi.strftime('%d/%m/%Y')
    #  lista_data_epi.append(data_epi)
    #  data_episodios = lista_data_epi
    #####
    #diretores = serie[''] #o ['directors'] nao retornou nada, talvez seja por episodio?
    #roteiristas
    escritores = serie['writer']
    lista_escri = []
    for escritor in escritores:
      lista_escri.append(escritor['name'])
    #try:
    #  produtoras = serie['production companies'] #seria serie['production companies']? acho que nao, pois aqui é produtora
    #except:
    #  produtoras = "erro"
    censura = serie['certificates']
    pais = serie['countries']
    try:
      idiomas = serie['languages']
    except KeyError:
      idiomas = ["NA"]

    linha = {'Titulo':titulo,
             'Check':checagem,
             'IMDBid':id,
             'Nota':nota,
             'Votos':votos,
             'Elenco': ", ".join(lista_elenco),#lista_elenco,
         'Gênero': ", ".join(genero),#genero,
         'Duração': ", ".join(duracao),#duracao,
         'Reviews':reviews,
         #'Prêmios':[premios],
         #'Sinopse':sinopse,
         'Ano':ano,
         'Data de estréia':data_estreia,
         #'Data dos episodios': ", ".join(data_episodios),#data_episodios,
         #'Diretores':[diretores],
         'Roteiristas':", ".join(lista_escri),
         #'Produtoras':",".join[produtoras],
         'Censura': ", ".join(censura),#censura,
         'País': ", ".join(pais),#pais,
         'Idiomas': ", ".join(idiomas),#idiomas,
         'Episodios':tamanho}    
    df = df.append(linha, ignore_index=True)   
  return df
```

# Pegando ids da nova amostra

```python
amostra = pd.read_csv("Amostra.csv")
```

```python
nomes = amostra['serie']
lista_ids = []
for nome in nomes:
    id="start"
    cont=0
    while(id=="start"):
        if (ia.search_movie(nome)[cont]['kind']=="tv series") or (ia.search_movie(nome)[cont]['kind']=="tv mini series"):
            id = ia.search_movie(nome)[cont].movieID
        cont=cont+1
    lista_ids.append(id)
```
